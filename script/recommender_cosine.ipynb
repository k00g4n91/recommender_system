```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import cosine_distances
import difflib
from recommender_module import get_sparse_matrix

movies= pd.read_csv('../data/movies.csv')
ratings = pd.read_csv('../data/ratings.csv')
movies_ratings = movies.merge(ratings,'inner')
dense_matrix = pd.read_csv('../data/dense_matrix.csv')
```

```python
sample = dense_matrix.iloc[:4, :3].copy()
sample
```

```python
#merge movies and ratings


# Sparse matrix function (0's not NaNs)
def get_sparse_matrix(dense_matrix: pd.DataFrame):
    return (dense_matrix
        .pivot_table(values='rating',columns='title', index='userId')
        .fillna(0)
    )

sparse_matrix = get_sparse_matrix(dense_matrix=movies_ratings)
sparse_matrix
```

```python
sample_similitude = pd.DataFrame(1 - cosine_distances(sparse_matrix), columns=sparse_matrix.index, index=sparse_matrix.index)
sample_similitude
```

```python
title = 'pulp fiction'
closest_match = difflib.get_close_matches(title, movies_ratings['title'], n=1).pop(0)
```

```python
weighted_ratings = sum(sparse_matrix[closest_match] * sample_similitude[1])
weighted_ratings
```

```python
#ponderate
round(weighted_ratings / (sum(sample_similitude[1]) - 1))
```
